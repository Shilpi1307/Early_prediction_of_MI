```python
kmms_dists = []
for k_means_model in wcss[:20]:
  # print(k_means_model.__dict__)

  kmm = k_means_model.__dict__
  n_clusters = kmm['n_clusters']
  cluster_centers = kmm['cluster_centers_']

  kmm["distances"] = [ [ 0 for i in range(n_clusters) ] for i in range(n_clusters) ]

  for i in range(n_clusters):
    for j in range(n_clusters):
      kmm["distances"][i][j] = np.linalg.norm(cluster_centers[i] - cluster_centers[j]).round(decimals=2)

  kmms_dists.append(kmm["distances"])
```

```python
for k, k_means_model in enumerate(wcss[:12]):
  plt.figure(figsize=(10, 10))
  G = nx.Graph()

  kmm = k_means_model.__dict__
  n_clusters = kmm['n_clusters']
  cluster_centers = kmm['cluster_centers_']

  for i in range(n_clusters):
    for j in range(n_clusters):
      if i < j:
        G.add_edge(str(i), str(j), weight=kmms_dists[k][i][j])

  pos = nx.spring_layout(G, seed=2)

  # nodes
  nx.draw_networkx_nodes(G, pos, node_size=700)

  # edges
  nx.draw_networkx_edges(G, pos, width=2)

  # node labels
  nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")

  # edge weight labels
  edge_labels = nx.get_edge_attributes(G, "weight")
  nx.draw_networkx_edge_labels(G, pos, edge_labels)

  ax = plt.gca()
  ax.margins(0.08)
  plt.axis("off")
  plt.tight_layout()
  plt.show()
```
